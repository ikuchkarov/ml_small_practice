## Iris dataset 
Begin with imports.

```python
import pandas as pd
import numpy as np
import scipy as sp
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

Load data from sklearn and split it to train and test.

```python
iris_data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_data['data'], iris_data['target'], random_state=44, train_size = 0.7)
```

Draw scatter matrix for train data.

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_data.feature_names)
pd.tools.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(16, 16), marker='o',
                           hist_kwds={'bins': 20}, s=20, alpha=.75);
```

We see, that one class is linearly separable from other.

Make baseline solution.

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score: {}".format(knn.score(X_test, y_test)))
```
